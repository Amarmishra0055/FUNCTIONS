6.   Explain the concept of generators in Python and how they are defined.
    -  In Python, a generator is a special type of iterator that allows you to create values one at a time as you need them, instead of creating all values at once and storing them in memory. This can save a lot of memory, especially when dealing with large data.You define a generator using a function, but instead of using return, you use the yield keyword. When the function is called, it doesn't run all the way through. Instead, it returns a generator object. When you call next() on the generator, it continues where it left off and yields the next value.

```python
#example

def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1


counter = count_up_to(5)


print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
```

7.  What are the advantages of using generators over regular functions?
   - In Python, generators have some key advantages over regular functions, especially when working with large data.Regular functions usually return all values at once, which means they store everything in memory, and this can use up a lot of memory if the dataset is large. Generators, on the other hand, only produce one value at a time when you need it. This way, they use less memory because they don’t store all the values at once.Generators are also useful because they only create values when you need them, which is known as lazy evaluation. This can be really helpful when you're working with data you don’t need all at once, or when the dataset is too large to store in memory.

```python
#wxample

def get_numbers(n):
    numbers = []
    for i in range(n):
        numbers.append(i)
    return numbers


def get_numbers_gen(n):
    for i in range(n):
        yield i
numbers_list = get_numbers(5)
print(numbers_list)  # [0, 1, 2, 3, 4]

numbers_gen = get_numbers_gen(5)
print(next(numbers_gen))  # Output: 0
print(next(numbers_gen))  # Output: 1
```

9.   Explain the purpose and usage of the `map()` function in Python.
   - In Python, the map() function is used to apply a specific function to all items in an iterable (like a list or tuple) and return a new iterable with the results. Instead of writing a loop to go through each item and apply a function, you can use map() to do it more easily and efficiently. Here's how it works: you give map() two things: the function you want to apply and the iterable (the list, tuple, etc.). It then applies the function to each element in the iterable and gives you a map object (which you can convert into a list or other types if you want).

```python
#example
def square(x):
    return x * x
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))
```

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - In Python, map(), reduce(), and filter() are all used to work with iterables like lists, but they serve different purposes.The map() function applies a given function to all items in an iterable and returns a new iterable with the results. For example, if you want to square each number in a list, you would use map() to apply the squaring function to each item.The reduce() function, which comes from the functools module, is used to apply a function to the items of an iterable in a cumulative way. It keeps applying the function to the items, using the result of the previous operation, until all items are processed. For example, you could use reduce() to find the sum or product of all the numbers in a list.The filter() function filters items from an iterable, keeping only those that meet a certain condition. It takes a function that returns True or False for each item, and only items that return True are kept in the result. For example, if you want to keep only the even numbers in a list, you would use filter().

```python
#example

def square(x):
    return x * x


numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))
```

```python
#1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

my_list = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)
```

```python
#4 Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(number, "is a prime number")
    else:
        print(number, "is not a prime number")
```

```python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.a if self.count > 1 else 0


for num in FibonacciIterator(10):
    print(num)
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power)
```

```python
# Create a Python program that uses `filter()` to remove all the vowels from a given string.


def remove_vowels(char):
    vowels = "aeiouAEIOU"
    return char not in vowels
input_string = "hello pw"

filtered_string = ''.join(filter(remove_vowels, input_string))
print(filtered_string)
```
